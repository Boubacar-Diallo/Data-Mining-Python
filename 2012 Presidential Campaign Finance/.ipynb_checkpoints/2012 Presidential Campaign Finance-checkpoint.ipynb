## 1) Basic campaign finance statistics
Output the basic statistics of Obama's and Romney's campaign finance (total non-zero/non-refund donations, minimum donation, maximum donation, average donation amount, median donation amount and standard deviation). Also maps the normalized sample contributions and z-score for the donation amounts of 20000 dollars, 10000 dollars, 5000 dollars, 1000 dollars, 500 dollars, 250 dollars, 100 dollars and 50 dollars. 

````python
# Campaign finance stats for Obama & Romney:
# Total positive donations, Minimum, Maximum, Mean, Median, Standard Deviation
# Normalized sample contributions & z-score for $20,000, $10,000, $5,000, $1,000, $500, $250, $100, $50
from collections import defaultdict
import csv, sys, datetime

#reader = csv.DictReader(open(sys.argv[1], 'r'))
reader = csv.DictReader(open("donations.txt", 'r'))

totalOb = 0
totalRom = 0
numcontOb = 0
numcontRom = 0
valuesOb = []
valuesRom = []
````

````python
for row in reader:
    name = row['cand_nm']
    datestr = row['contb_receipt_dt']
#    amount = float(row['contb_receipt_amt'])
    if 'Obama' in name:
        amountOb = float(row['contb_receipt_amt'])
        # Exclude negative and $0 donations from all calculations 
        if amountOb > 0:
            totalOb += amountOb
            valuesOb.append(amountOb)
            numcontOb += 1
    if 'Romney' in name:
        amountRom = float(row['contb_receipt_amt'])
        # Exclude negative and $0 donations from all calculations
        if amountRom > 0:
            totalRom += amountRom
            valuesRom.append(amountRom)
            numcontRom += 1
````

Aggregate Obama's numbers here:

````python
meanOb = totalOb/numcontOb
medianOb = sorted(valuesOb)[len(valuesOb)/2]
minValOb = min(valuesOb)
maxValOb = max(valuesOb)

def sd(valuesOb,meanOb):
    sizeOb = numcontOb
    
    # Find deviation from mean for each value and square it
    squaresOfDevOb = [(v-meanOb)**2 for v in valuesOb]
    
    # Sum the squared deviation of each value
    sumOfDevOb = 0.0
    for v in squaresOfDevOb:
        sumOfDevOb += v
    
    # Divide the sum of squared deviations by one less than the total count of values. Then find the square root.
    stdDevOb = (sumOfDevOb/(sizeOb-1))**0.5
    
    return stdDevOb

stdDevOb = sd(valuesOb,meanOb)

def minmax_normalizeOb(value):
    # Takes a donation amount and returns a normalized value between 0-1.
    normOb = ((value-minValOb)/(maxValOb-minValOb))*(1-0)+0

    return normOb

def zscoreOb(value):
    # Calculates the z-score for a given value, mean and standard deviation
    zScoreOb = (value-meanOb)/stdDevOb
    
    return zScoreOb
````

Aggregate Romney's numbers here:

````python
meanRom = totalRom/numcontRom
medianRom = sorted(valuesRom)[len(valuesRom)/2]
minValRom = min(valuesRom)
maxValRom = max(valuesRom)

def sd(valuesRom,meanRom):
    sizeRom = numcontRom
    
    # Find deviation from mean for each value and square it
    squaresOfDevRom = [(v-meanRom)**2 for v in valuesRom]
    
    # Sum the squared deviation of each value
    sumOfDevRom = 0.0
    for v in squaresOfDevRom:
        sumOfDevRom += v
    
    # Divide the sum of squared deviations by one less than the total count of values. Then find the square root.
    stdDevRom = (sumOfDevRom/(sizeRom-1))**0.5
    
    return stdDevRom

stdDevRom = sd(valuesRom,meanRom)

def minmax_normalizeRom(value):
    # Takes a donation amount and returns a normalized value between 0-1. 
    normRom = ((value-minValRom)/(maxValRom-minValRom))*(1-0)+0

    return normRom

def zscoreRom(value):
    # Calculates the z-score for a given value, mean and standard deviation
    zScoreRom = (value-meanRom)/stdDevRom
    return zScoreRom
````

````python
print "Obama's campaign finance stats:\n-----------------------------------"
print "Total positive donations: $%s" % totalOb
print "Numbers of non-refund/non-zero donations: %s" % numcontOb
print "Minimum positive donation: $%s" %  minValOb
print "Maximum positive donation: $%s" % maxValOb
print "Mean positive donation: $%s" % meanOb
print "Median donation: $%s" % medianOb 
print "Standard Deviation of donations: $%s" % stdDevOb
print "Min-max normalized values: %r" % map(minmax_normalizeOb, [20000, 10000, 5000, 1000, 500, 250, 100, 50])
print "Z-Scores: %r \n" % map(zscoreOb, [20000, 10000, 5000, 1000, 500, 250, 100, 50])

print "Romney's campaign finance stats:\n-----------------------------------"
print "Total positive donations: $%s" % totalRom
print "Numbers of non-refund/non-zero donations: %s" % numcontRom
print "Minimum positive donation: $%s" %  minValRom
print "Maximum positive donation: $%s" % maxValRom
print "Mean positive donation: $%s" % meanRom
print "Median donation: $%s" % medianRom 
print "Standard Deviation of donations: $%s" % stdDevRom
print "Min-max normalized values: %r" % map(minmax_normalizeRom, [20000, 10000, 5000, 1000, 500, 250, 100, 50])
print "Z-Scores: %r \n" % map(zscoreRom, [20000, 10000, 5000, 1000, 500, 250, 100, 50])
````

## 2) Compare Obama's and Romney's campaign donations by states

````python
from collections import defaultdict
import csv, sys, datetime

reader = csv.DictReader(open("donations.txt", 'r'))

obamadonations = defaultdict(lambda:0)
romneydonations = defaultdict(lambda:0)
````

````python
# States to exclude from Romney's calculation
stateRom = set(('33', '46', '48', '7', '77', '8', '84', '91', 'AA', 'AE', 'AP', 
                'AS', 'FF', 'GE', 'GU', 'HA', 'MP', 'PE', 'PR', 'TE', 'UK', 'VI', 'XX'))

# States to exclude from Obama's calculation
stateOb = set(('AA', 'AB', 'AE', 'AM', 'AP', 'AS', 'BC', 'BR', 'DI', 'FM', 'GU', 'IT', 'MB', 
               'MH', 'MP', 'NA', 'NO', 'NS', 'ON', 'PE', 'PR', 'PW', 'QC', 'QU', 'SA', 'VI', 
               'ZZ'))
````

````python
for row in reader:
    name = row['cand_nm']
    datestr = row['contb_receipt_dt']
    state = row['contbr_st']
    amount = float(row['contb_receipt_amt'])
    
    if 'Obama' in name:
        if  amount > 0:
            if any(c in state for c in stateOb):
                continue
            else:
                obamadonations[state] += amount
                #print name + '\t' + state + '\t' + str(obamadonations[state])
    if 'Romney' in name:
        if  amount > 0:
            if any(c in state for c in stateRom):
                continue
            else:
                romneydonations[state] += amount
                #print name + '\t' + state + '\t' + str(romneydonations[state])
````

Print and output the result to `states.txt`.

````python
OUT = open("states.txt", "w")
OUT.write("State" + "\t" + "Obama's donations ($)" + "\t" + "Romney's donations ($)" + "\n")
print "%s\t%s\t%s" % ("State","Obama's donations ($)","Romney's donations ($)")
for state, amount in sorted(obamadonations.items(), key=lambda (key,val): key):
    OUT.write("%s\t%s\t%s\n" % (state,obamadonations[state],romneydonations[state]))
    print "%s\t%s\t%s" % (state,obamadonations[state],romneydonations[state])
OUT.close()
````

According to the FEC, Mitt Romney received higher total positive campaign donations (679994941.51 dollars) than Barack Obama (558359124.9  dollars). Across 50 states and Washington D.C., Obama out raised Romney in only 14 states that are Democratic stronghold: California, District of Columbia, Delaware, Hawaii, Illinois, Massachusetts, Maryland, Maine, New Mexico, New York, Oregon, Rhode Island, Vermont, Washington. 

The 3 states where Obama received the highest amount of donations were Illinois (96.25 million dollars), California (93.64 million dollars) and New York (51.73 million dollars). The 3 states where Romney received the highest amount of donations were California (79.06 million dollars), Texas (70.05 million dollars) and Florida (58.76 million dollars).

## 3) Compare Obama's and Romney's monthly campaign donations between March 2011 and December 2012

````python
%matplotlib inline
from collections import defaultdict
import matplotlib.pyplot as plt
import csv, sys, datetime

reader = csv.DictReader(open("donations.txt", 'r'))

obamadonations = defaultdict(lambda:0)
romneydonations = defaultdict(lambda:0)
````

````python
for row in reader:
    name = row['cand_nm']
    datestr = row['contb_receipt_dt']
    amount = float(row['contb_receipt_amt'])
    date = datetime.datetime.strptime(datestr, '%d-%b-%y')

    if 'Obama' in name:
        obamadonations[date] += amount
    if 'Romney' in name:
        romneydonations[date] += amount
````

````python
fig = plt.figure(figsize=(12,6))    # create a 12-inch x 6-inch figure
sorted_by_dateob = sorted(obamadonations.items(), key=lambda (key,val): key)
sorted_by_datemc = sorted(romneydonations.items(), key=lambda (key,val): key)
xs1,ys1 = zip(*sorted_by_dateob)
xs2,ys2 = zip(*sorted_by_datemc)
plt.plot(xs1, ys1, label="Obama's Donations")
plt.plot(xs2, ys2, label="Romney's Donations")
plt.legend(loc='upper center', ncol = 4)
#plt.savefig('obamaRomney.png', format='png')
````

Mitt Romney received donations early beginning in May 2011 as he prepared to enter the Republican presidential primaries (February 1 2012 - April 1 2012). His donations increased significantly in April 2012 after he clinched the Republican nomination. Obama, on the other hand, began his re-election campaign in January 2012 and he was formally nominated as President at the Democratic National Convention in Charlotte, North Carolina, on September 6 2012. Obama finalized his campaign on November 6 2012 as he was re-elected with 51% popular vote and 332 electorate votes.

## 4) Compare Obama's and Romney's cumulative campaign donations between March 2011 and December 2012

````python
import csv,sys,datetime,collections
import itertools
import matplotlib.pyplot as plt

reader = csv.DictReader(open("donations.txt", 'r'))

totaldonations = collections.defaultdict(list)
````

````python
for row in reader:
    name = row['cand_nm']
    datestr = row['contb_receipt_dt']
    amount = float(row['contb_receipt_amt'])
    date = datetime.datetime.strptime(datestr, '%d-%b-%y')
    
    if 'Obama' in name or 'Romney' in name:
        totaldonations[name].append((date, amount))

campaigntotals = dict([(name, sum(map(lambda p:p[1], val))) for name, val
                   in totaldonations.iteritems()])
````

````python
fig = plt.figure(figsize=(12,6))    # create a 12-inch x 6-inch figure

idx = 0
# Obama's and Romney's cumulative donations
for name, monies in totaldonations.iteritems():
    monies.sort(key=lambda pair: pair[0])
    i = itertools.groupby(monies, key=lambda p: p[0])
    monies = map(lambda (key, pairs): (key, sum([float(pair[1]) for pair in pairs])), i)
    total = 0
    newmonies = []
    for pair in monies:
        total += pair[1]
        newmonies.append((pair[0], total ))
    monies = newmonies

    xs,ys = zip(*monies)
    plt.plot(xs, ys, label = name + "'s donations")
    idx += 1
plt.legend(loc='upper center', ncol = 4)
#plt.savefig('cumulative donations.png', format='png')
````

Mitt Romney and Barack Obama had similar cumulative donation amounts until September 2012, after Obama and Biden were formally nominated for President and Vice President at the 2012 Democratic National Convention. The increased donations to Mitt Romney likely signaled efforts by his contributors to reach out to potential voters through ground work and television commercials.

## 5) Compares Obama's and Romney's cumulative campaign reattributions between March 2011 and December 2012


As it turns out, "redesignations" and "reattributions" of donations are [normal](http://www.fec.gov/pages/brochures/contrib.shtml#Excessive_Contributions). For instance, if a donation by person A is excessive, the part that exceeds the limits can be "reattributed" to person B, meaning that person B donated the rest to the campaign. Alternatively, the excess amount can be redesignated to another campaign in the same party. So a donation to Romney could be redesignated to a poor republican in Nebraska. However, "Reattribution to spouse" can be suspicious. A potential theory is that CEOs and wealthy contributors use it as a tactic to [hide campaign contributions](http://dailycaller.com/2011/07/06/huntsmans-website-double-the-donation-with-one-for-your-spouse/). For example, a CEO could donate money, then reattribute them to the donor's spouse. Then the spouse will donate that amount to the election candidate. In this way, a casual browser will find it difficult to notice that a the candidate is supoorted by a company's CEOs.

````python
import csv,sys,datetime,collections
import itertools
import matplotlib.pyplot as plt

reader = csv.DictReader(open("donations.txt", 'r'))

totalreattributions = collections.defaultdict(list)
````

````python
for row in reader:
    name = row['cand_nm']
    datestr = row['contb_receipt_dt']
    amount = float(row['contb_receipt_amt'])
    date = datetime.datetime.strptime(datestr, '%d-%b-%y')
    reason = row['receipt_desc']
    
    if amount < 0 and 'REATTRIBUTION' in reason:
        if 'Obama' in name or 'Romney' in name:
            totalreattributions[name].append((date, -amount))

candreattributions = dict([(name, sum(map(lambda p:p[1], val))) for name, val
                   in totalreattributions.iteritems()])
````

````python
fig = plt.figure(figsize=(12,6))	# create a 12-inch x 6-inch figure

idx = 0
# Obama's and McCain's cumulative reattributions
for name, monies in totalreattributions.iteritems():
    monies.sort(key=lambda pair: pair[0])
    i = itertools.groupby(monies, key=lambda p: p[0])
    monies = map(lambda (key, pairs): (key, sum([float(pair[1]) for pair in pairs])), i)
    total = 0
    newmonies = []
    for pair in monies:
        total += pair[1]
        newmonies.append((pair[0], total ))
    monies = newmonies

    xs,ys = zip(*monies)
    plt.plot(xs, ys, label = name + "'s reattributions")
    idx += 1
plt.legend(loc='upper center', ncol = 4)
#plt.savefig('cumulative reattributions.png', format='png')
````

The campaign data indicates that Obama did not receive any reattributed donation but Romney received about 4.1 million dollars in reattribution especially between June 2012 and October 2012 when the presidential election race heated up.

## 7) Compare frequency of Obama's and Romney's campaign donations between -3000 dollars and 3000 dollars with a histogram and a boxplot

````python
for row in reader:
    name = row['cand_nm']
    amount = float(row['contb_receipt_amt'])
    candtomoney[name].append(amount)

obama = candtomoney["Obama, Barack"]
romney = candtomoney["Romney, Mitt"]

increment = 100
width=50
obama_bucketed = map(lambda amount: amount - amount%increment, obama)
romney_bucketed = map(lambda amount: amount - amount%increment, romney)

obama_hist = Counter(obama_bucketed)
romney_hist = Counter(romney_bucketed)
minamount = min(min(obama), min(romney))
maxamount = max(max(obama), min(romney))
print "min/max", minamount, maxamount
buckets = range(int(minamount), int(maxamount+1), increment)
````

````python
fig = plt.figure(figsize=(12,6))
subplot = fig.add_subplot(111)
subplot.bar(obama_hist.keys(), obama_hist.values(), color='b', edgecolor='b', label="Barack Obama")#, s=1)
subplot.bar(romney_hist.keys(), romney_hist.values(), color='r', edgecolor='r', label="Mitt Romney")#, s=1)

# We can't see anything without a log scale subplot.set_yscale('log')
# there are anamolous donations greater than 100,000 that we don't want to see
subplot.set_xlim((-3000, 3000))
subplot.legend()
#plt.savefig('obama romney histogram.png', format='png')
````

## 8) Perform ttests on Obama's and Romney's campaign donations data

We perform a Welch's T-test on the data to determine if the difference between Romney's and Obama's average campaign contribution is significant. The reported p-value is within rounding error of 0, which is statistically significant.

````python
# Part I
# Run Welch's T-test on donations.txt. Is the difference between Romney and Obama's average 
# compaign contribution significant?
import csv,sys,datetime,collections
import numpy
import scipy.stats
import welchttest

reader = csv.DictReader(open("donations.txt", 'r'))
idx = 0

candtomoney = collections.defaultdict(list)
````

````python
for row in reader:
    name = row['cand_nm']
    amount = float(row['contb_receipt_amt'])
    candtomoney[name].append(amount)

obama = candtomoney["Obama, Barack"]
romney = candtomoney["Romney, Mitt"]
````

````python
print "Welch's T-Test p-value:", welchttest.ttest(obama, romney)
````

Because the Welch's T-test makes no assumption about the size of the dataset or its variances, we have likely violated the assumption that the campaign finance data is normally distributed. We conduct the Shapiro-Wilk test to check if the data is actually normal for both presidential candidates. In this case, the test calculates a p-value and tells us that it is not normally distributed if the p-value <0.05. 

Obama has a Shapiro-Wilks p-value of 0.000577, which indicates that we have violated the normality assumption of the Welch's T-test.

````python
print "Obama's Shapiro-Wilks p-value", scipy.stats.shapiro(obama)
print "Romney's Shapiro-Wilks p-value", scipy.stats.shapiro(romney)
````

Because T-Tests are resilient to breaking of the normality assumption and there are nonparametric equivalents that don't make normality assumptions, we run the Mann-Whitney U nonparametric T-test to determine on the campaign dataset. The reported p-value is about 0, so the result is still statistically significant.

````python
print "mann-whitney U", scipy.stats.mannwhitneyu(obama, romney)
````
