## 1. Gather the number of employees on each campus (Ann Arbor, Dearnborn, Flint)

```python
import re

# Compiling regular expressions
a2_re = re.compile(r"ANN-ARBOR")      # match to a string that contains "ANN-ARBOR"
dborn_re = re.compile(r"DEARBORN")    # match to a string that contains "DEARBORN"
flnt_re = re.compile(r"FLINT")        # match to a string that contains "FLINT"

# Open the salaries data file
salary_file = open("salaries2012.txt", "r")

headers = salary_file.readline()
```

```python
# Initialize variables
prevname = ""

a2_count = 0
dborncount = 0
flntcount = 0
```

```python
for line in salary_file.readlines():
    line.rstrip()
    (campus, name, title, dept, ftr, rpt_period, fte, gf) = line.split('\t')
    if name == prevname:
        continue
    else:
        prevname = name
    
    if not flnt_re.search(campus):
        if not dborn_re.search(campus):
            a2_count += 1
        else:
            dborncount += 1
    else: 
        flntcount += 1
```

```python
print "Total number of employees in Ann Arbor in 2012 = %d" % a2_count
print "Total number of employees in Dearborn in 2012 = %d" % dborncount
print "Total number of employees in Flint in 2012 = %d" % flntcount
```

## 2. Sort the job titles by the number of people employed under that job title

```python
from collections import Counter

salary_file = open("salaries2012.txt", "r")

headers = salary_file.readline()

titles = []
```

```python
for line in salary_file.readlines():
    line.rstrip()
    (campus, name, title, dept, ftr, rpt_period, fte, gf) = line.split('\t')
    titles.append(title)
```

```python
print "\nMost common university job titles in 2012"
print "-------------------------------------------"
i = 0
for i, (title, titlecount) in enumerate(Counter(titles).most_common()):
    i += 1
    print "%d. %s\t%s" % (i, title, titlecount)
```

## 3. Sort the job titles by the average salary for that title

```python
import collections

Job = collections.namedtuple('Job', ['title', 'salary'])

salaries = {}

salary_file = open("salaries2012.txt", "r")

headers = salary_file.readline()
```

```python
for line in salary_file.readlines():
    line.rstrip()
    (campus, name, title, dept, ftr, rpt_period, fte, gf) = line.split('\t')
    ftr = float(ftr.replace('"', '').replace(',', ''))
    if title in salaries:
        salaries[title].append(ftr)
    else:
        salaries[title] = [ftr]
```

```python
average = lambda alist: sum(alist)/len(alist)
average_salaries = [Job(title, average(salaries[title])) for title in salaries]

print "\nJob titles with the highest average salary\n------------------------------"
sorted_salaries = sorted(average_salaries, key=lambda job: job.salary, reverse=True)
for i, j in enumerate(sorted_salaries):
    print '%i. %s \t$%g' % (i + 1, j.title, j.salary)
```

## 4. Create a tag cloud of the First names for professors in Ann Arbor

```python
import re

# Compiling regular expressions
# match to a string that contains "ANN-ARBOR"
a2_re = re.compile(r"ANN-ARBOR")

# Match to a full professor, the '^' character denotes the beginning of a line
# in text: PROFESSOR, Professor and PROFESSOR EMERITUS/A
prof_re = re.compile(r"^PROFESSOR", re.I | re.X)

salary_file = open("salaries2012.txt", "r")

headers = salary_file.readline()

# Compile a regular expression that find the characters after the first
# comma and labels them as "firstname"
firstname_re = re.compile(r"[^,]+,(?P<firstname>\w+|.)", re.I)

prevname = ""
profs = {}
```

```python
for line in salary_file.readlines():
    line.rstrip()
    (campus, name, title, dept, ftr, rpt_period, fte, gf) = line.split('\t')
    
    # Some individuals are listed under multiple departments and 
    # and we skip two names in a row that are the same
    if name == prevname:
        continue
    else:
        prevname = name
    if not a2_re.search(campus):
        #print "Not in Ann Arbor: %s" % line
        continue
    else:
        #print "In Ann Arbor: %s" % line
        firstname = firstname_re.match(name).group('firstname')
    
    if prof_re.search(title):
        #print "%s" % line
        #print "Prof: %s\t%s" % (firstname, title)
        # Trying to count the names of the profs (so either sets default as 0, or returns its value) then add 1
        profs[firstname] = profs.setdefault(firstname,0)+1
```

```python
# sort by name frequency
print "%s\t%s" % ("Professor name", "Name count")
print "-----------------------------"
for name, namecount in sorted(profs.iteritems(), key=lambda (k,v): (v,k), reverse=True):
    print "%s\t%s" % (name, namecount)
```
